```python
from queue import Queue
import random

class Request:
  def __init__(self, name, id):
    self.name = name
    self.id  = id #порядковий номер заявки

class Work_with_requests :
  def __init__(self):
    self.requests = Queue()

  def generate_request(self, request):
    self.requests.put(request)
    print("створено нову завку")

  def process_request(self):
    if not self.requests.empty():
      current_request = self.requests.get()
      print(f"опрацьовуємо завку {current_request.name} з id: {current_request.id}")
    else:
      print("Черга порожня")
      

# Створюємо процедуру роботи з заявками
w_w_r = Work_with_requests()
x = 'Y' 
print('Процес створення обробки заявок')
i = 0
while x == 'Y':
    # Додаємо запити
    var = random.randint(1, 2)
    print(var)
    if var == 1:
        w_w_r.generate_request(Request(f"Заявка-{i}", i))
        i+=1
    else :
# опрацьовуємо заявку
        w_w_r.process_request()
    print("Для завершення введіть N, для продовження введіть Y")
    print("Ваша команда: ")
    x = input()
```
